```python
import firebase_admin
from firebase_admin import credentials,firestore

cred = credentials.Certificate("ServiceAccountKey.json")
firebase_admin.initialize_app(cred)
db = firestore.client()
def save_to_firestore(collection_name, data):
    try:
        for key, value in data.items():
            doc_ref = db.collection(collection_name).document(str(key))
            doc_ref.set(value)
        print("Data successfully written to Firestore")
    except Exception as e:
        print(f"Error writing to Firestore: {e}")
```

```python
import json
with open('reordered.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
save_to_firestore('Lemma', data)
```
