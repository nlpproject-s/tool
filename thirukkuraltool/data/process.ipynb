```python
import csv
import json
result_dict = {}
with open('output_data - Data.csv', 'r', encoding='utf-8') as csvfile:
    csvreader = csv.reader(csvfile)
    for row in csvreader:
        if len(row) != 3:
            continue  
        main_key=row[0] 
        meaning=row[1]   
        num=row[2]
        try:
            num= int(num)
        except ValueError:
            continue  
        if main_key not in result_dict:
            result_dict[main_key]={}
        result_dict[main_key][num]=meaning
json_output = json.dumps(result_dict, ensure_ascii=False, indent=4)
print(json_output)
with open('output.json', 'w', encoding='utf-8') as jsonfile:
    jsonfile.write(json_output)
```
